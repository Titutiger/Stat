```python
from rapidfuzz import fuzz as rapidfuzz_fuzz, process as rapidfuzz_process
import pandas as pd
import numpy as np
from timeit import default_timer as timer
from functools import wraps
import ipywidgets as widgets
import sys
import os

from all_functions import df_functions


# Add project root to sys.path to import our functions list if needed
# sys.path.append(os.path.abspath("../../../"))
def print_timing(f):
    @wraps(f)
    def inner(*args, **kwargs):
        t0 = timer()
        try:
            return f(*args, **kwargs)
        finally:
            # print(f"\n{f.__name__} finished in {1e3*(timer()-t0):.1f} ms")
            pass

    return inner


# We can import it or just paste the data here for the notebook to be self-contained

data = df_functions
df = pd.DataFrame(data)
function_names = df["name"].values


@print_timing
def search_functions(input_text):
    if not input_text:
        print("Type a function name to see its documentation...")
        return

    results = rapidfuzz_process.extract(input_text, function_names, limit=5, scorer=rapidfuzz_fuzz.WRatio)

    print(f"{'Function':<25} | {'Documentation'}")
    print("-" * 80)

    for name, score, index in results:
        description = df.iloc[index]["description"]
        print(f"{name:<25} | {description}")


widgets.interact(search_functions, input_text="");
```
