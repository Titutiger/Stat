## 1. Defining a stat obj

A stat obj is defined using the `represent` method. It is the recommended way for creating a stat obj.

```python
import pandas as pd

import src.stat as stat

pd_df = pd.read_csv('../examples/assets/medical_trials.csv')

# definition:
df = stat.represent(pd_df)
df
```

If one tries to print the `df`, it will return: `Stat(tag='dataframe', theme='default')`.

The tagging is down by checking if the valued passed is a `pd.dataframe` or a `np.ndarray`

---

## 2. Representing a stat obj

A stat obj can be displayed using the `show()` method:

```python
df.show()
```

The `show()` parameter accepts 2 arguments, the first is the tile of your data contained and the second is the theme it is to be displayed in:

```python
df.show('Medical Trials', 'ocean')
```

The `filter_types()` method is used to filter numeric and non-numeric data from your container:

```python
df.filter_types('n').show('', 'cyan') # n -> numeric
```

```python
df.filter_types('nn').show('','forest') # nn -> non-numeric
```

If I want to check the count of all elements in `Group`, then:

```python
df.frequencies('Group').show()
```

```python
# specific count:
df.count_value('Group', 'Control')
```
