## 1. Defining a stat obj

A stat obj is defined using the `represent` method. It is the recommended way for creating a stat obj.

```python
import pandas as pd

import src.stat as stat

pd_df = pd.read_csv('../examples/assets/insurance.csv')

# definition:
df = stat.represent(pd_df)
df
```

If one tries to print the `df`, it will return: `Stat(tag='dataframe', theme='default')`.

The tagging is down by checking if the valued passed is a `pd.dataframe` or a `np.ndarray`

---

## 2. Representing a stat obj

A stat obj can be displayed using the `show()` method:

```python
df.show()
```

The `show()` parameter accepts many arguments, out of which the first is the tile of your data contained and the second is the theme it is to be displayed in:

```python
df.show('Medical Trials', 'ocean')
```

```python
df.show(max_rows=5)
```

```python
df.show(max_rows=0) # for only columns
```

```python
df.show(max_rows=5, width='*') # changes the width of the table either according to a specified value or the terminal size.
```

```python
df.filter_types('n').show('', 'cyan', max_rows=5) # n -> numeric
```

```python
df.filter_types('nn').show('','forest', max_rows=5) # nn -> non-numeric
```
