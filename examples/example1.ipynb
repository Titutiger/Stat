# Example1
We are an HR analyst looking at a small tech startup's payroll.

```python
import numpy as np
import pandas as pd
from src.stat.core import Stat
```

The following is the data that we have:

```python
data = {
    'Employee': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Age': [28, 32, np.nan, 41, 55],               # Charlie's age is missing!
    'Years_Exp': [2, 5, 4, 10, 20],
    'Salary': [60_000, 75_000, 72_000, 90_000, 1_200_000] # Eve makes 1.2 million!
}
df = pd.DataFrame(data)
```

Firstly, let's initialize Stat:

```python
hr_stats = Stat(df)

# dimensions:
hr_stats.shape
```

```python
# average salary:
hr_stats.mean(series='Salary')


# Skewed heavily by CEO's salary
```

```python
# The true middle is the median:
hr_stats.median('Salary')
```

Measuring salary equality (spread):

```python
hr_stats.std(series='Salary') # blown out of proportion due to CEO
```

```python
# Therefore, median absolute deviation is used:
hr_stats.mad(series='Salary')
```

```python
'''
The iqr shows the gap between the 25th and 75th percentile:
'''
hr_stats.iqr(series='Salary')
```

```python
'''NaN values:'''
hr_stats.mean(series='Age')
```

```python
hr_stats.summary()
```
