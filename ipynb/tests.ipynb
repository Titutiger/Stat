````python
data = Stat([4, 8, 6, 5, 3, 8])
````

Basic descriptions:

````python
# mean:
round(data.mean(), 4)
````

Here, `data.mean()` takes 1 argument `method: str` wherein it's default value is `arithmatic`.
`data.mean()` can take: `(a)rithmatic`, `(g)eometric` and `(h)armonic`.

- a = ari = arithmetic
- g = geo = geometric
- h = har = harmonic

````python
# mean (geometric & harmonic)
print(round(data.mean('g'), 4))
print(round(data.mean('h'), 4))
````

With the same way:

````python
print(round(data.median(), 4))
print(round(data.mode(), 4))
print(round(data.range(), 4))
````

Herein, `median()` has a parameter `return_index` which is set to `False`.

If turned `True`, it returns:

````python
data.median(return_index=True)
````
