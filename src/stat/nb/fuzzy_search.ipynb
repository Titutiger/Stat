```python
def print_timing(f):
    @wraps(f)
    def inner(*args, **kwargs):
        t0 = timer()
        try:
            return f(*args, **kwargs)
        finally:
            # print(f"\n{f.__name__} finished in {1e3*(timer()-t0):.1f} ms")
            pass
    return inner
```

```python
@print_timing
def search_functions(input_text):
    if not input_text: 
        print("Type a function name to see its documentation...")
        return

    results = rapidfuzz_process.extract(input_text, function_names, limit=5, scorer=rapidfuzz_fuzz.WRatio)
    
    print(f"{'Function':<25} | {'Documentation'}")
    print("-" * 80)
    
    for name, score, index in results:
        description = df.iloc[index]["description"]
        print(f"{name:<25} | {description}")

widgets.interact(search_functions, input_text="");
```
